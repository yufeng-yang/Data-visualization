# Step2: Data Preparation

First observe whether their distribution is Gaussian distribution, and then choose the normalization method.

```python
import scipy.stats as stats
z_scores_ghgs = stats.zscore(df_final['mean_ghgs'])
# Generating Artificially Normally Distributed Data for mean_ghgs
theoretical_qq_ghgs = np.sort(stats.norm.rvs(size=len(df_final['mean_ghgs'])))

z_scores_bio = stats.zscore(df_final['mean_bio'])
# Generating Artificially Normally Distributed Data for mean_bio
theoretical_qq_bio = np.sort(stats.norm.rvs(size=len(df_final['mean_bio'])))

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6)) 

stats.probplot(df_final['mean_ghgs'], dist="norm", plot=ax1)
ax1.set_title('Q-Q Plot of mean_ghgs')

stats.probplot(df_final['mean_bio'], dist="norm", plot=ax2)
ax2.set_title('Q-Q Plot of mean_bio')

plt.tight_layout()
plt.show()
```

**None of them are normal distributions, so max-min will be used for normalization instead of the standardization.**  
Max-min normalizes the specified features and converts them to the 0-10 range

```python
features_to_normalize = ["mean_ghgs", "mean_land","mean_eut", "mean_bio", "mean_watuse","mean_acid"]
for feature in features_to_normalize:
    min_value = df_final[feature].min()
    max_value = df_final[feature].max()
    df_final[feature] = ((df_final[feature] - min_value) / (max_value - min_value) * 10)

df_selected = df_final.copy()
```

```python
import pandas as pd
features_to_average = ["mean_ghgs", "mean_land", "mean_eut", "mean_bio", "mean_watuse", "mean_acid"]

info_columns = ["diet_group"]

grouped_features = df_selected.groupby('diet_group')[features_to_average].mean().round(2)

grouped_info = df_selected.groupby('diet_group')[info_columns].first()

final_grouped_df = pd.concat([grouped_info, grouped_features], axis=1)

final_grouped_df

# 如果需要保存到CSV
# final_grouped_df.to_csv('Diet_Impact_on_Environment.csv', index=True)
```

## Calculated Relative Percentage Change

```python
veggie = final_grouped_df[final_grouped_df['diet_group'] == 'veggie'].iloc[0, 1:]  # Get row for veggie and drop the diet_group column
meat100 = final_grouped_df[final_grouped_df['diet_group'] == 'meat100'].iloc[0, 1:]  # Get row for meat100 and drop the diet_group column
# Calculate Relative Percentage Change for each category
rpc = (veggie - meat100) / meat100 * 100
rpc

```

```python
for col in rpc.index:
    final_grouped_df[col + '_RPC'] = (final_grouped_df[col] - meat100[col]) / meat100[col] * 100
final_grouped_df
```
